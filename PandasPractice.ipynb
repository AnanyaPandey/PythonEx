```python
#Common Pokemon ; Pokemon who is not Legendary
commonpoke = pokemon[pokemon['Legendary']==False]
commonpoke['Legendary'].sum()
```

```python
# pokemon who are not mega in their name 
nonmegapoke = pokemon[~pokemon['Name'].str.contains('Mega')]
```

```python
#check how many pokemon are there for every number
checkpoke = pokemon['Number'].groupby(pokemon['Number']).count().sort_values(ascending=False)
checkpoke
```

```python
# consider same name as duplicates ; kepe first and drop others
noduppoke = pokemon.drop_duplicates('Number', keep='first',inplace = False)
```

```python
len( noduppoke['Number'].unique())
```

```python
len( pokemon['Number'].unique())
```

```python
# Which pokemon type is the strongest and which the weakest on average? (according to total stats)

strongestavg = noduppoke[['Type 1','Total']].groupby(noduppoke['Type 1']).mean().sort_values(by='Total',ascending=False)
strongestavg
```

```python
test = noduppoke[['Type 1','Total']].groupby(noduppoke['Type 1'])
test
```

```python

strongmax = noduppoke[['Type 1','Total','Name']].sort_values(by='Total', ascending=False).groupby('Type 1')
strongmax
```

```python
top5_poke = noduppoke[['Name', 'Total']].sort_values(by='Total', ascending=False)[0:5]
top5_poke
```

```python
generationpoke = noduppoke[['Generation','Total']].groupby(noduppoke['Total']).mean().sort_values(by='Total',ascending=False)
```
