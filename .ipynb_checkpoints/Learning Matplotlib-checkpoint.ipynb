```python
x1 = [1,2,7]
x2=[2,4,8]
y1 = [6,7,11]
y2 = [3,4,10]

plt.plot(x1,y1, label = "Year 2017", linewidth = 3 )
plt.plot(x2,y2, label = "Year 2018", linewidth = 3 )

plt.title("Year Comparison")
plt.legend()
plt.grid(True,color='grey')
```

```python
K18 = [[2,7,4],[7,11,9]]
K17 = [[3,6,5],[9,5,10]]

plt.bar(K18[0],K18[1], label = "Example Year 18 ", color='red')
plt.bar(K17[0],K17[1], label = "Example year 17", color='green')

plt.legend()
```
