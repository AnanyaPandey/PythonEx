```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,10,20)
y=x**2
```

```python
x
```

```python
plt.plot(x,y,'r')
plt.xlabel("X Value")
plt.ylabel("Y Value")
```

```python
plt.plot(y,x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r')
plt.subplot(2,1,2)
plt.plot(y,x,'r--')
```

```python
# Object Oriented method 
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Some Function of Square');
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,'r')
axes[1].plot(x,y*y,'b')
plt.tight_layout()
```
