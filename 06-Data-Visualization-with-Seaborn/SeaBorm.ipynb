```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],bins=30)
```

```python
print(sns.jointplot(x='tip',y='total_bill',data=tips, kind='hex'))
```

```python
sns.pairplot(tips[['total_bill','tip']])
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='tip',data=tips, hue='smoker')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='smoker')
```
